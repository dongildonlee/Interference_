```python
import os
import re

# specify the directory path
dir_path = os.path.realpath('../../data/raw_response/')

print("Directory being searched:", dir_path)

# iterate over files in the directory
for filename in os.listdir(dir_path):
    print("Found file:", filename)
    if filename.endswith(".mat"):
        print("yes")
        # parse relu value
        relu_value = int(re.search(r'relu(\d+)', filename).group(1))

        # check relu condition
        if relu_value == 5:
            # remove '_test' from the filename
            new_filename = filename.replace('_test', '')
            
            # rename the file
            os.rename(os.path.join(dir_path, filename), 
                      os.path.join(dir_path, new_filename))
```
